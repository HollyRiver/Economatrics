### EDA

> 데이터의 특징과 구조 파악
>
> 전처리 과정, 자료를 완벽하게 분석할 수 있도록 검증 및 수정(올바른 자료인지, 이상치가 잘못 기록된 것은 아닌지 등등...)

있는 그대로의 데이터 분석.


### 데이터 분석의 4가지 주제

1. 저항성
   
   저항성이 있는 중위수를 선호

> 일부 자료의 부실한 측정, 이상치의 존재로 인해 평균은 저항성이 없음.

2. 잔차

   각 개별 관측치가 주 경향으로부터 얼마나 벗어났는지 확인

3. 재표현

   대칭화, 산포를 비슷하게 만듦(로그, 제곱, 역수변환 등)

4. 현시성

   그래픽으로 표현하여 자료를 탐색


 저항성이 있는 통계량을 쓰고, 잔차를 써서 데이터를 분석하고, 데이터를 대칭화하고 산포를 비슷하게 만들고, 그래픽으로 표현하여 자료를 탐색하자.


### 탐색적 자료분석의 기본철학

$$모형 ↔ 데이터 ↔ 분석$$

위 과정을 계속해서 반복하는 것, 따라서 이 과정을 최소화하기 위해 우선적으로 데이터를 탐색하는 과정이 요구된다.

> ex) 이상치를 넣고 분석했는데, 이후 그 사실을 알아챈 경우 분석을 다시 해야 한다.
>
> 따라서 전처리 과정이 아주 중요하다.

### 탐색적 자료분석 성공사례?

>
