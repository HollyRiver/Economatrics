```r
n = 50000
p = 0.99

x = seq(round(n*p-2*sqrt(n*p*(1-p))), round(n*p+2*sqrt(n*p*(1-p))), 1)

plot(x, dbinom(x, n, p), xlim = c(n*p-2*sqrt(n*p*(1-p)), n*p+2*sqrt(n*p*(1-p))), type = "l")
curve(dnorm(x, n*p, sqrt(n*p*(1-p))), add = T, col = "red", lty = 2)
```
