```r
x = rnorm(1000)
qqnorm(x)
qqline(x)
```

```r
shapiro.test(x)  ## 기각불가
```

```r
y = runif(100, 0, 1)
qqnorm(y)
qqline(y)
```

```r
shapiro.test(y)  ## 기각, 정규분포 아님
```

`-` 기타 다른 분포에 대해 qqplot을 그릴 경우

```r
qqnorm(c(rnorm(25, 130, 15), rnorm(25, 70, 15)))
qqline(c(rnorm(25, 130, 15), rnorm(25, 70, 15)))
```

```r
qqnorm(rnorm(25, 130, 15) + rnorm(25, 70, 15))
qqline(rnorm(25, 130, 15) + rnorm(25, 70, 15))
```

`-` 특이값이 있는 정규분포

```r
qqnorm(c(25, 175, rnorm(38, 100, 15)))
qqline(c(25, 175, rnorm(38, 100, 15)))
```

* 혼합정규분포에서는 직선이 두 개 형성된다.
* 특이점이 있는 경우 직선을 벗어난 점이 몇 존재한다.
* 꼬리가 짧은 분포일 경우 양 끝이 눌려진다.
* 꼬리가 긴 분포의 경우 양 끝이 늘어난다.

* 왼쪽으로 기울어지면 역 U자를, 오른쪽으로 기울어지면 U자를 따르는 그래프가 나온다.

`-` 지수분포의 확률플롯

```r
leukemia = c(1,1,2,2,3,4,4,5,5,8,8,8,8,11,11,12,12,15,17,22,23)

n = length(leukemia)
p = (seq(1:n)-0.5)/n
x = -log(1-p)
y = leukemia
plot(x~y, main = "QQ plot for exponential dist")
```
