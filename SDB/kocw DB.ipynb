제거할 정보 : 제공기관명 & 제공기관코드 & 데이터기준일자

```python
kocw.drop(["제공기관명", "제공기관코드", "데이터기준일자"], axis = 1)
```

강의학기 : 강의년도와 강의학기로 나뉘어짐

등록일자 : 등록년도, 등록월, 등록일로 나뉘어짐

```python
kocw.drop(["제공기관명", "제공기관코드", "데이터기준일자"], axis = 1).reset_index().rename({"index" : "강의번호"}, axis = 1)
```

```python
len(set(kocw.강의URL주소.str.split("=").str[-1])) ## url request가 모두 다른 것을 확인
```

```python
kocw_data = kocw.drop(["제공기관명", "제공기관코드", "데이터기준일자"], axis = 1)\
.assign(강의번호 = lambda _df : _df.강의URL주소.str.split("=").str[-1].astype("int"))\
.set_index("강의번호").reset_index()  ## 인덱스 열을 앞으로 오도록 변경
```
