```python
import pandas as pd
```

```python
kocw = pd.read_excel("/content/kocw_data.xls", header=1)
```

```python
region = pd.read_csv("/content/전국대학및전문대학정보표준데이터.csv",  encoding='euc-kr')
```

```python
#전국대학및전문대학정보표준데이터에서 소재지 데이터 추출
#kocw데이터에 국립 단어가 빠져있는 경우가 많아 전국대학및전문대학정보표준데이터 '경상국립대' 제외 나머지 국립제거
#소재지 시도 데이터 추가(도 시, 시 구)
main_campus_filtered = region[region['본분교구분명'] == '본교'][['학교명', '소재지도로명주소']]
main_campus_filtered['학교명'] = main_campus_filtered['학교명'].apply(lambda x: x if x == '경상국립대학교' else x.replace('국립', ''))
main_campus_filtered['소재지도로명주소'] = main_campus_filtered['소재지도로명주소'].str.split().str[:2].str.join(' ')
region_list = list(main_campus_filtered['소재지도로명주소'])
trns = {main_campus_filtered['학교명'].iloc[i]: region_list[i] for i in range(len(main_campus_filtered))}
kocw = kocw.assign(소재지=kocw['제공대학명'].map(trns))
```

2. 강의코드 인덱스 새기고 테이블 분리

5. 담당 관계에 담당교수코드와 강의코드를 매핑(방법을 좀 생각해봐야 함.)

```python
## line1 : 교육 중분류명이 없는 튜플 제거
## line2~3 : 강의코드 인덱스를 강의URL주소에서 따오고 정수형으로 변경. 추출된 열은 제거
featured_kocw = kocw.loc[kocw["교육 중분류명"].isna().map(lambda x : False if x else True), :]\
.assign(강의코드 = lambda _df : _df.강의URL주소.str.split("=").str[-1].astype("int"))\
.set_index("강의코드").reset_index().drop("강의URL주소", axis = 1).reset_index(drop = True)
featured_kocw
```

```python
## 테이블 분리
professor = featured_kocw[["교수명", "교육 대분류명", "제공대학명", "강의코드"]].rename({"교수명":"담당교수명", "교육 대분류명":"담당분야"}, axis = 1)  ## 담당교수
university = featured_kocw[["제공대학명"]]  ## 제공대학
course = featured_kocw[["강의코드", "공개강의명", "교육 중분류명", "강의학기", "제공대학명"]].rename({"공개강의명":"강의명", "교육 중분류명":"교육분류"}, axis = 1)  ## 강의
charge = featured_kocw[["강의코드", "교수명"]].rename({"교수명":"담당교수명"}, axis = 1)  ## 담당
```

```python
## professor 테이블에서 "+"로 구분되는 교수명 행으로 분리, 이후 담당교수명과 담당분야, 제공대학명으로 구성되는 테이블에 대하여 중복되는 값을 없게 하고 인덱스로 담당교수코드 할당.
## {담당교수명, 담당분야, 제공대학명}으로 구성된 데이터와 담당교수코드 두 개의 열로 이뤄진 테이블을 만들고, map({})을 이용하여 담당 관계에 담당교수코드와 강의코드를 매핑
professor
```
