### 데이터베이스 설계 단계

1. 요구사항 분석

> 요구사항을 분석하여 그 다음 단계인 E-R 다이어그램을 구성하는 데 도움이 되도록 정리한다.
>
> 사용자의 요구사항을 정확히 분석해야 재수정, 재개발의 비용을 절감할 수 있다.

개체 외 그에 필요한 값을 명명, 데이터를 식별할 수 있는 속성을 고려

2. 개념적 설계

> 개체와 속성, 관계를 추출하고 E-R 다이어그램을 제작한다.
>
> 개체와 속성은 명사 위주로 추출하고, 개체 간 관계는 동사 위주로 추출한다.

필수와 선택, 전체 참여나 종속성 여부까지 고려

3. 논리적 설계

> 개념적 구조를 기반으로 논리적 구조를 설계한다.
>
> E-R 다이어그램으로부터 릴레이션 스키마를 생성한다.

기본규칙 5가지 이용

> 1) 모든 개체의 릴레이션 변환
> 2) 다대다 관계의 릴레이션 변환
> 3) 일대다 관계의 외래키 표현
>
> > 일반 개체의 경우 다쪽의 개체에 일쪽 릴레이션의 기본키와 관계에 포함된 속성을 붙인다.
> >
> > 약한 개체가 존재하면 관계에 속성정보가 있기 어렵다. 그냥 이어주면 된다.
> 
> 4) 일대일 관계의 외래키 표현
>
> > 중복을 없애기 위해 필수참여 개체에만 외래키를 붙일 수도 있다.
> > 
> > 둘다 필수 참여 개체인 경우 릴레이션을 합친다.
> 
> 5) 다중값 속성의 릴레이션 변환 : 튜플이 더 많아짐

4. 물리적 설계

> 구체적인 내부 저장 구조, 접근 경로, 저장 레코드, 인덱스 설계, 탐색 기법 등을 정의한다.

5. 구현

> DBMS를 이용하여 실제 데이터베이스를 구축한다.

### 정규화

> 속성 간 연관성이 부분적임에도 하나의 릴레이션에 둠으로써 데이터 중복이 발생하고, 이에 따라 삽입/수정/삭제 연산에서의 부작용인 이상현상이 발생한다. 하나의 릴레이션에는 강한 연관성을 가진 것들이 모여있도록 하여 이상현상을 제거하는 과정이다.
