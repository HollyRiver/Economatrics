### 1. 데이터베이스 설계

네, 1조 발표 시작하도록 하겠습니다. 저는 데이터베이스 설계 과정을 발표할 강신성입니다.

저희 1조는 KOCW라는 대학 원격 강좌 플랫폼의 강의 데이터를 주제로 DB를 구성하였는데요. 진행의 순서부터 말씀드리자면, 먼저 사용한 데이터를 설명해드리고, 다음으로 요구사항 분석, 개념적 설계, 논리적 설계까지 설명해드린 뒤, 실제 R과 Python으로 구축된 데이터베이스의 시연 결과를 보여드리는 것으로 마무리하겠습니다.

사용한 데이터입니다. 저희가 가져온 데이터는 공공 데이터 포털에 한국교육학술정보원이 제공한 KOCW 강의 데이터로 한 개의 테이블만 존재합니다. 실제 데이터를 보시면 이와 같은데, 16822개의 강의와 12개의 변수를 가지고 있습니다. 저희는 이 중 유의미한 정보만 사용하기 위해 16814개의 강의 8개의 변수만을 택하였고, 추가적으로 정보 추출을 위하여 대학이 위치한 소재지 정보와 사립/국립에 대한 정보를 직접 손으로 수집하여 추가하였습니다.

요구사항 분석입니다. 개체에 대한 문장에서 관계에 대한 문장 순서대로 구성하여 보았는데요. 개체에 대한 문장부터 차례대로 읽어보자면, 제공대학이 존재하려면 대학명과 소재할 위치가 있어야 하고, 설립 구분 정보를 포함하고 대학명으로 식별한다. 강의에 대한 강의코드, 강의명, 교육분류, 강의학기 및 강의년도에 대한 정보를 유지하고, 강의는 강의코드로 식별한다. 담당교수는 플랫폼에 가입할 때 이름과 소속분야를 입력해야 하며, 각 교수는 담당교수코드를 부여하여 식별한다.

그리고 관계인데요, 제공대학에는 여러 담당교수가 소속될 수 있지만, 한 명의 교수는 여러 대학에 소속될 수 없다. 여기서 일대다 관계가 나올 수 있었고,

담당교수는 여러 강의를 담당할 수 있고, 여러 교수가 강의 하나를 담당할 수 있다. 여기서는 다대다 관계가 나올 수 있었습니다.

제공대학은 여러 강의를 제공할 수 있지만, 여러 대학이 하나의 강의를 공동으로 제공하지 않는다에서 일대다 관계가 나올 수 있었고, 제공대학이 강의를 제공하면 제공일자 정보를 유지해야 합니다.

개념적 설계 과정입니다. 요구사항 분석을 기반으로 개체와 속성, 관계를 추출할 수 있었는데요. 이를 기반으로 E-R 다이어그램을 만들어낼 수 있었습니다. 보시면, 모든 개체와 관계가 이중 도형으로 그려졌고, 관계를 나타내는 선은 이중선인 것을 보실 수 있습니다. 이는 DB가 한 개의 테이블에서 비롯되었기 때문에 모든 개체와 관계가 완전종속이라는 점에서 유도되는 결과입니다.

다음은 논리적 설계 과정입니다. E-R 다이어그램에서 릴레이션 스키마를 뽑아낼 때, 3개의 개체와 한개의 다대다 관계가 존재하므로 총 4개의 릴레이션을 유도할 수 있었는데요. 이를 정리하자면 이렇게 표시할 수 있습니다. 여기서 담당 관계의 경우 외래키로만 구성되어 있으며, 두 외래키는 이차원의 기본키가 됩니다.

뽑아낸 릴레이션별로 함수종속 관계를 도시화했습니다. 제공대학 릴레이션부터 보시면 기본키인 제공대학명이 대학소재지와 설립구분을 종속시키고 있고요, 담당교수 릴레이션의 경우 기본키인 교수코드가 나머지 속성들을 종속시키고 있습니다.

강의 릴레이션의 경우에는 이상현상이 발생할 수 있는 일부 튜플만 따로 추출해서 보여드렸는데요. 보시는 바와 같이 강의명이 같으면 교육분류가 동일하여 기본키 이외 속성 간 함수종속관계가 있는 것을 알 수 있습니다. 이를 그림으로 나타내면 강의코드와 교육분류 간 이행적 함수종속관계가 있는 것을 알 수 있습니다. 담당 릴레이션의 경우 새로 담당코드 열을 만들어 기본키로 삼으면 함수종속관계를 만들 수 있긴 한데, 불필요한 정보라 생각하여 생성하지 않았습니다.

그래서 정규화 계획을 세워보았습니다. 보이스-코드 정규화까지 진행함에 있어 아까 보여드린 강의 테이블의 이행적 함수종속만 문제였는데, 이는 강의 테이블을 강의와 강의교육분류 둘로 분리하여 해소할 수 있었습니다. 그리하여 최종적으로 다섯 개의 릴레이션을 뽑아낼 수 있었고, 이를 기반으로 데이터베이스를 구축하게 되었습니다.
