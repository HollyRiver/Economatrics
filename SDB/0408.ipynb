`-` group-wise statistic / select

> 그룹별 통계량

```r
e = starwars %>% group_by(species) %>% summarise(
    n = n(), mass = mean(mass, na.rm = TRUE)) %>% filter(n > 1, mass > 60)
e
```

> `group_by()` : 판다스 groupby()라고 보면 됨.

`-` Join

```r
set.seed(222)
d1 = data.frame(id = rep(1:10), c = rpois(10*1, 2))
d2 = data.frame(id = 2:11, d = rpois(10,4))
```

```r
inner_join(d1, d2, by = "id")
```

```r
full_join(d1, d2, by = "id")
```

```r
right_join(d1, d2, by = "id")
```

```r
left_join(d1, d2, by = "id")
```

```r
merge(d1, d2, by = "id", all.x = T, all.y = T)
```

> `pd.merge(df1, df2, by = "id", how = "right, left, ...")`

`-` 일반 집합 연산자 : union, intersection, setdiff

```r
set.seed(222)
d1 = data.frame(id = rep(1:5), c = rpois(5*1,2))
d2 = data.frame(id = c(1,2,4,5,6), c = c(d1[1:2, 2], rpois(3,4)))
```

```r
union(d1, d2)
```

> 다르기만 하면 다 더해짐

```r
intersect(d1, d2)
```

```r
setdiff(d1, d2)
```

```r
setdiff(d2, d1)
```

`-` cartesian product

```r
x = rnorm(100); v = rnorm(100); y = rnorm(100); rbinom(100, 1, 0.5)

d1 <- data.frame(x=x, v=v, c=v)
expand(data.frame(y=y, z=z), d1, nesting(y, z))  ## 앞의 것들을 고정시키고 y_z를 바꾸는 식으로 변경
```
